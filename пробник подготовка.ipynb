## Линейная регрессия

**Постройте модель линейной регрессии, которая показывает, как потребление вина на душу населения (Wine_PerCapita) влияет на уровень счастья в стране (HappinessScore). Укажите в качестве ответа значение свободного коэффициента. Ответ округлите до двух знаков.** 

**Пример ответа: 0.00**

```python
#идём по инструкции из легальной шпоры
import statsmodels.api as sm
X = sm.add_constant(df['Wine_PerCapita'])
Y = df['HappinessScore']
model = sm.OLS(Y, X).fit()
model.params
```

## MAE, MSE, R2

**Постройте модель линейной регрессии, которая показывает, как потребление вина на душу населения (Wine_PerCapita) влияет на уровень счастья в стране (HappinessScore). Укажите в качестве ответа значение коэфициента детерминации (R2) . Ответ округлите до двух знаков.** 


```python
from statsmodels.tools.eval_measures import mse, meanabs
```

```python
Y_pred = model.predict(X)
mse1 = mse(Y,Y_pred)
print(mse1)
mae1 = meanabs(Y,Y_pred)
print(mae1)
r2 = model.rsquared
print(r2)
```

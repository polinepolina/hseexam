## Линейная регрессия
**Постройте модель линейной регрессии, которая показывает, как среднее время приема у врача (Avg Time to Patients(mins)) влияет на процент пациентов, удовлетворенных работой врача (Patient Satisfaction Rate). В качестве ответа укажите произведение полученных параметров модели. Ответ округлите до двух знаков.**

Пример ответа: 0.00

```python
#X = Avg Time to Patients(mins)
#Y = Patient Satisfaction Rate
import statsmodels.api as sm
X = sm.add_constant(df['Avg Time to Patients(mins)'])
Y = df['Patient Satisfaction Rate']
model = sm.OLS(Y, X).fit()
model.params
```

```python
a =97.857801*-0.085059
print(a)
```

```python
#MAE
Y_pred = model.predict(X)
from statsmodels.tools.eval_measures import mse, meanabs
MAE = meanabs(Y,Y_pred)
print(MAE)
```
